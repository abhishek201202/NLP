```python
import pandas as pd
#reading the data
data = pd.read_csv("spam.csv", encoding = "ISO-8859-1")
data.head()
```

```python
#expanding the dispay of text sms column
pd.set_option('display.max_colwidth', -1)
#using only v1 and v2 column
data= data [['v1','v2']]
data.head()
```

```python
#checking the count of the dependent variable
data['v1'].value_counts()
```

# punctuation removal

```python
#library that contains punctuation
import string
string.punctuation
```

```python
#defining the function to remove punctuation
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation])
    return punctuationfree
#storing the puntuation free text
data['clean_msg'] = data['v2'].apply(lambda x:remove_punctuation(x))
data.head()
```

```python
data['msg_lower']= data['clean_msg'].apply(lambda x: x.lower())
```

```python
data
```

# Tokenization

```python
#defining function for tokenization
import re
from nltk.tokenize import word_tokenize
def tokenization(text):
    tokens = word_tokenize(text)
    return tokens
#applying function to the column
data['msg_tokenied']= data['msg_lower'].apply(lambda x: tokenization(x))
```

```python
data
```

# Spelling Correction

```python
from autocorrect import Speller
spell = Speller(lang = "en")
#defining the function to remove stopwords from tokenized text
def spelling_correction(text):
    corrected =[]
    for i in text:
        corrected.append(spell(i))
    return corrected
```

```python
#applying the function
data['msg_lower'] = data['msg_tokenied'].apply(lambda x:spelling_correction(x))
```

```python
data
```

# Stopwords Removal

```python
#importing nlp library
import nltk
#Stop words present in the library
stopwords = nltk.corpus.stopwords.words('english')
stopwords[0:10]
['i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', "you're"]
```

```python
#defining the function to remove stopwords from tokenized text
def remove_stopwords(text):
    output= [i for i in text if i not in stopwords]
    return output
```

```python
#applying the function
data['no_stopwords']= data['msg_tokenied'].apply(lambda x:remove_stopwords(x))
```

```python
data
```

# Stemming

```python
#importing the Stemming function from nltk library
from nltk.stem.porter import PorterStemmer
#defining the object for stemming
porter_stemmer = PorterStemmer()
```

```python
from nltk.tokenize import word_tokenize
#defining a function for stemming
def stemming(text):
    temp = word_tokenize(str(text))
    stem_text = [porter_stemmer.stem(word) for word in temp]
    return stem_text
data['msg_stemmed']=data['msg_tokenied'].apply(lambda x: stemming(x))
```

```python
data
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
#defining the object for Lemmatization
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
#defining the function for lemmatization
def lemmatizer(text):
    temp = word_tokenize(str(text))
    lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in temp]
    return lemm_text
data['msg_lemmatized']=data['no_stopwords'].apply(lambda x:lemmatizer(x))
```

```python
data
```

# Remove frequent words

```python
def remove_fre_words(lemmas):
    frequency = {}
    for lemma in lemmas:
        if lemma in frequency:
            frequency[lemma] += 1
        else:
            frequency[lemma] = 1

    filtered_text = []
    for w in lemmas:
        if frequency[w] <= 3:
            filtered_text.append(w);
    return filtered_text
```

```python
data['remove_frequent_words']=data['msg_lemmatized'].apply(lambda x:remove_fre_words(x))
```
