# Gutenberg Corpus

```python
import nltk
from nltk.corpus import stopwords
from nltk.probability import FreqDist
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt

# Download necessary Data
nltk.download('gutenberg')
nltk.download('punkt')
nltk.download('stopwords')

# Load austen-emma.txt from gutenberg Corpus
text = nltk.corpus.gutenberg.words('austen-emma.txt')

# Load all English Stop Words in a set
stopset = set(stopwords.words('english'))

# Get all tokens from text after removing stop words
tokenized_word = word_tokenize(' '.join([word for word in text if word.isalnum() and word not in stopset]))

print(tokenized_word)
print(len(tokenized_word))
```

```python
# Plot of the Freq Distribution of the words
fdist = FreqDist(tokenized_word)
fdist.plot(30, cumulative = False)
plt.show()
```
